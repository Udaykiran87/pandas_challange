```python
url1 = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
```

```python
price_dolar = df['item_price']
```

```python
df['raw_price'] = price_dolar.str.replace('$', '', regex=False).astype(float)
```

```python
df['total_price'] = df['quantity']*df['raw_price']
```
